```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel sheet
excel_file = r'\sales_funnel_data.xlsx'  # Replace with your Excel file
df = pd.read_excel(excel_file)

# Ensure your Excel sheet has columns like 'Stage' and 'Count'
# Example data:
#   Stage           Count
#   Prospects       1000
#   Leads           600
#   Opportunities   300
#   Customers       150

# Plotting the sales funnel
plt.figure(figsize=(10, 6))

# Bar plot for the counts at each stage
plt.bar(df['Stage'], df['Count'], color=['gray', 'blue', 'orange', 'green'])

# Adding labels and title
plt.xlabel('Sales Funnel Stages')
plt.ylabel('Number of Accounts')
plt.title('Sales Funnel Visualization')

# Adding data labels
for i, count in enumerate(df['Count']):
    plt.text(i, count + 10, str(count), ha='center', va='bottom')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Value': np.random.randn(100)
})

# Set the style for seaborn
sns.set(style="whitegrid")



# 6. Heatmap (for correlation matrix)
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
