```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from Excel file
excel_file_path = ('/kaggle/input/tngdba/BarPlot.xlsx')  # Replace with the actual path to your Excel file
df = pd.read_excel(excel_file_path)

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', data=df)
plt.title('Bar Plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Read data from Excel file
excel_file_path = ('/kaggle/input/tngdba/BoxPlot.xlsx')  # Replace with the actual path to your Excel file
df = pd.read_excel(excel_file_path)

# Set the style for seaborn
sns.set(style="whitegrid")

# 3. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Box Plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
excel_file_path = ('/kaggle/input/tngdba/LinePlot.xlsx')
df = pd.read_excel(excel_file_path)

# Line plot
plt.plot(df['Date'], df['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Line Plot')
plt.show()
```
